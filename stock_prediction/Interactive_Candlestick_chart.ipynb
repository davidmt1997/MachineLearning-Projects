```
import plotly.graph_objects as go
import pandas as pd
```

```
import pandas_datareader as web
```

```
#Get the stock quote
df = web.DataReader('AMZN', data_source='yahoo', start='2019-05-01', end='2020-05-31')
df
```

```
# Create an interactive candlestick chart
figure = go.Figure (
    data = [
            go.Candlestick(
                x = df.index,
                low = df['Low'],
                high = df['High'],
                close = df['Close'],
                open = df['Open'],
                increasing_line_color = 'green',
                decreasing_line_color = 'red'
            )
    ]
)

figure.show()
```

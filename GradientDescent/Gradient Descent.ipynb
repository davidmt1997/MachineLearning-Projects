## Gradient Descent implementation from scratch

```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
# anonymous function that will be used to do the gradient descent
func = lambda th: np.sin(1/2 * th[0]**2 - 1/4 * th[1]**2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1])

# creating the vectors X and Y with 100 numbers from -2 to 2
# vector Z for the height of the function, which is what we want to minimize
resolution = 100
_X = np.linspace(-2, 2, resolution)
_Y = np.linspace(-2, 2, resolution)
_Z = np.zeros((resolution, resolution))

#populate vector Z
for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x, y])

# visualizing the depth of the function
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()

# generating a vector of 2 numbers from -2 to 2
Theta = np.random.rand(2) * 4 -2

_T = np.copy(Theta)

h = 0.001
lr = 0.001

plt.plot(Theta[0], Theta[1], "o", color="white")

# vector used to minimize the parameters and the function
grad = np.zeros(2)

# iterate 1000 times to find the minimum value in the function
for _ in range(10000):
    for it, th in enumerate(Theta):

        _T = np.copy(Theta)
        
        # add a small number 
        _T[it] = _T[it] + h

        # calculating the ratio of the difference between the two functions
        deriv = (func(_T) - func(Theta))/h

        #save the calculated value, which is the slope of the axis in it
        grad[it] = deriv

    # "take the step" part in which we move in the "downwards" direction
    Theta = Theta - lr * grad    
    
    #print(func(Theta))

    if(_ % 100 == 0):
        plt.plot(Theta[0], Theta[1], ".", color="red")
        
plt.plot(Theta[0], Theta[1], "o", color="green")
plt.show()
```
